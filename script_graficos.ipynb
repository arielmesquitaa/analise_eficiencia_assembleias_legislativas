#### 1. Cria indicadores secundários

```python
list(df_indicadores)
```

```python
## Ajustando nomes de coluna
df_indicadores.columns = ['estados',
 'qtde_leis',
 'populacao',
 'IDH',
 'PIB',
 'gasto_orcamento',
 'gasto_pessoal_ativo',
 'gasto_pessoal_ativo_LRF',
 'proporcao_gasto_pessoal_RCL',
 'qtde_sessoes_plenarias',
 'gasto_previsto_cota_por_deputado',
 'qtde_deputados']
```

```python
dias_uteis_2016 = 252 ## Considerei os 366 dias de 2016 menos os feriados nacionais.
df_indicadores["proporcao_dias_uteis_sessoes"] = df_indicadores["qtde_sessoes_plenarias"]/dias_uteis_2016
df_indicadores["gasto_por_deputado"] = df_indicadores["gasto_orcamento"]/df_indicadores["qtde_deputados"]
df_indicadores["gasto_por_habitante"] = df_indicadores["gasto_orcamento"]/df_indicadores["populacao"]
df_indicadores["gasto_por_lei"] = df_indicadores["gasto_orcamento"]/df_indicadores["qtde_leis"]
df_indicadores["gasto_pessoal_ativo_deputado"] = df_indicadores["gasto_pessoal_ativo"]/df_indicadores["qtde_deputados"]
df_indicadores["leis_por_deputado"] = df_indicadores["qtde_leis"]/df_indicadores["qtde_deputados"]
df_indicadores["proporcao_gasto_orcamento_no_PIB"] = df_indicadores["gasto_orcamento"]/df_indicadores["PIB"]
df_indicadores["verba_gabinete_IDH"] = df_indicadores["gasto_previsto_cota_por_deputado"]/df_indicadores["IDH"]
df_indicadores["PIB_per_capita"] = df_indicadores["PIB"]/df_indicadores["populacao"]
```

```python
df_indicadores
```

```python
## Indicadores normalizados pro ranking de eficiência
lista_indicadores = ["leis_por_deputado", "proporcao_dias_uteis_sessoes", "gasto_por_habitante", "gasto_por_lei", "proporcao_gasto_orcamento_no_PIB", "verba_gabinete_IDH", "gasto_pessoal_ativo_deputado", "proporcao_gasto_pessoal_RCL"]
for i in lista_indicadores:
    nm_new_col = i+"_norm"
    max_col = df_indicadores[i].max()
    min_col = df_indicadores[i].min()
    mean_col = df_indicadores[i].mean()
    # funcao
    def func_norm(x):
        if pd.isnull(x) == True:
            x = mean_col
            y = (x - min_col)/(max_col - min_col)
        else:
            y = (x - min_col)/(max_col - min_col)
        return y
    df_indicadores[nm_new_col] = df_indicadores[i].map(func_norm)
```
